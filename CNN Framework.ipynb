```python
from skimage.io import imread
from tensorflow.keras.preprocessing import image
import cv2 as cv
import matplotlib.pyplot as plt

def load_image_files(train_data):
  images = []
  Y = []
  for obj in train_data.values:
    #print(obj[0])
    img = cv.imread(obj[0])
    #img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    img_pred = cv.resize(img, (64, 64), interpolation=cv.INTER_AREA)
    img_pred = image.img_to_array(img_pred)
    img_pred = img_pred / 255
    #print(img_pred.shape)
    #if img_pred.shape[0]*img_pred.shape[1]*img_pred.shape[2] != 12288:
      #print("Hayo")
    img_pred = img_pred.reshape(img_pred.shape[0]*img_pred.shape[1]*img_pred.shape[2])
    images.append(img_pred)
    Y.append(obj[1])
    #print(img_pred.shape)
  X = np.array(images)
  Y = np.array(Y)
  print("Y Shape: ", Y.shape)
  print("X Shape: ", X.shape)
  return X, Y
```

```python
import matplotlib.pyplot as plt

def draw_performance(CV, VALIDATION_ACCURACY, VALIDATION_PRECISION, VALIDATION_RECALL, VALIDATION_F1SCORE, algorithm):

  fig, axs = plt.subplots(2,2,constrained_layout=True, figsize=(15,8))

  axs[0,0].plot(CV, VALIDATION_ACCURACY, color='red', marker='o')
  axs[0,0].set_title('Accuracy ', algorithm)
  axs[0,0].set_xlabel('Cross Validation')
  axs[0,0].set_ylabel('Score')
  axs[0,0].set_ylim((0.5, 1))
  axs[0,0].grid()

  axs[0,1].plot(CV, VALIDATION_PRECISION, color='green', marker='o')
  axs[0,1].set_title('Precision ', algorithm)
  axs[0,1].set_xlabel('Cross Validation')
  axs[0,1].set_ylabel('Score')
  axs[0,1].set_ylim((0.5, 1))
  axs[0,1].grid()

  axs[1,0].plot(CV, VALIDATION_RECALL, color='blue', marker='o')
  axs[1,0].set_title('Recall ', algorithm)
  axs[1,0].set_xlabel('Cross Validation')
  axs[1,0].set_ylabel('Score')
  axs[1,0].set_ylim((0.5, 1))
  axs[1,0].grid()

  axs[1,1].plot(CV, VALIDATION_F1SCORE, color='purple', marker='o')
  axs[1,1].set_title('F1 Score ', algorithm)
  axs[1,1].set_xlabel('Cross Validation')
  axs[1,1].set_ylabel('Score')
  axs[1,1].set_ylim((0.5, 1))
  axs[1,1].grid()
```
